```python
import pandas 
import numpy as np
import matplotlib as mp
# import the data into a pandas dataframe
file="nj_teachers_salaries_2016.csv"
teachers=pandas.read_csv(file)
```

```python
#check out what the data looks like 
teachers.head()
```

```python
#find all the employees that are working in Edison Township
criteria=teachers['district']=='Edison Twp'
edison=teachers[criteria]
#check that this is what should be in the table
edison.head()
```

```python
#mean salary of edison teachers
mean=edison['salary'].mean()
print("Mean: ")
print(mean)
#median salary of edison teachers
median=edison['salary'].median()
print("Median:")
print(median)
```

```python
#group the data by schools in the distict
schools=edison.groupby('school')
schools.size()
```

```python
#scatter plot of teachers salaries from the two high schools 
ax1=schools.get_group("John P. Stevens High School").plot(x="experience_total",y="salary",kind="scatter",color="g")
schools.get_group("Edison High School").plot(x="experience_total",y="salary",kind="scatter",ax=ax1,color="r")
mp.pyplot.title("Teacher Salary Comparison: JPS vs EHS")
# mean salary and experience of the teachers in each high school 
print("JP STEVENS:")
sal1=schools.get_group("John P. Stevens High School")['salary'].mean()
print(sal1)
print("Experience: ")
exp1=schools.get_group("John P. Stevens High School")['experience_total'].mean()
print(exp1)
print("Median: ")
med1=schools.get_group("John P. Stevens High School")['salary'].median()
print(med1)
print("\n")
print("EDISON HIGH SCHOOL")
sal2=schools.get_group("Edison High School")['salary'].mean()
print(sal2)
print("Experience: ")
exp2=schools.get_group("Edison High School")['experience_total'].mean()
print(exp2)
print("Median: ")
med2=schools.get_group("Edison High School")['salary'].median()
print(med2)
```
